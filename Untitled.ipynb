```python
def corelation(col1,col2):
    xdas = col1.mean()
    ydas = col2.mean()
    cov = np.sum((col1-xdas)*(col2-ydas))/(col1.shape[0])
    r = cov/(col1.std()*col2.std())
    return r
logAvgMonthlyHours = raw_data[raw_data.left == 1]['average_monthly_hours']
satisfaction_levelLeft = raw_data[raw_data.left == 1]['satisfaction_level']
corelation(satisfaction_levelLeft,logAvgMonthlyHours)
```

```python
data.drop(columns = ['Department'],inplace=True,axis = 1)
columns = ['satisfaction_level', 'last_evaluation', 'number_project',
       'average_monthly_hours', 'centroidDis','time_spend_company', 'Work_accident',
       'promotion_last_5years', 'DepartmentGroup_1',
       'DepartmentGroup_2', 'DepartmentGroup_3', 'salary_1', 'salary_2',
       'salary_3','left']
data = data[columns]
```

```python
binaryValColStart = 5
continousData = data.iloc[:,:binaryValColStart]
binaryData = data.iloc[:,binaryValColStart:]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.utils import shuffle
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.feature_selection import f_regression
```

```python
from sklearn.metrics import confusion_matrix
```

```python
continousDataScaler = StandardScaler()
continousDataScaled = pd.DataFrame(continousDataScaler.fit_transform(continousData),columns =continousData.columns)
```

```python
scaledData = pd.concat([continousDataScaled,binaryData],axis = 1)
```

```python
left_size = scaledData['left'].sum()
```

```python
balancedData = shuffle(scaledData,random_state = randomState).copy()
left_size = scaledData['left'].sum()
unleft_size = 0
for i in range(balancedData.shape[0]):
    if scaledData.iloc[i].left == 0 :
        unleft_size+=1
        if unleft_size > left_size:
            balancedData.drop([i],inplace = True,axis = 0)
        
```

```python
balancedData.left.sum()/balancedData.shape[0],balancedData.shape[0]
```

```python
yScaledData = balancedData['left'].copy()
xScaledData = balancedData.drop(['left'],axis = 1)
xtrain,xtest,ytrain,ytest = train_test_split(xScaledData,yScaledData,test_size = 0.1,random_state = randomState)
```

```python
model = LogisticRegression()
model.fit(X = xtrain,y = ytrain)
```

```python
model.intercept_
```

```python
Rsqrt = model.score(xtrain,ytrain)
n = xtrain.shape[0]
p = xtrain.shape[1]
adjRsqrt = 1-(1-Rsqrt)*((n-1)/(n-p-1))
print(Rsqrt,adjRsqrt)
```

```python
p_value = f_regression(xtrain,ytrain)[1]
```

```python
summary = pd.DataFrame({'featuers':xScaledData.columns.copy()})
summary['coef'] = model.coef_.reshape(-1,1)
summary['odds'] = np.exp(summary['coef'])
summary['p_value'] = p_value.round(3)
summary['mean'] = data.values.mean(axis = 0)[:-1]
```

```python
summary.sort_values('odds',ascending = False)
```

```python
prediction = model.predict(xtest)
confMatrix = confusion_matrix(y_true = ytest,y_pred = prediction)
```

```python
print(confMatrix)
confMatrix = confMatrix.reshape(-1)
```

```python
print(f"""
accuracy          = {round((confMatrix[0]+confMatrix[3])/confMatrix.sum(),4)}
misclassification = {round((confMatrix[1]+confMatrix[2])/confMatrix.sum(),4)}
sensitivity       = {round(confMatrix[3]/(confMatrix[3]+confMatrix[2]),4)}
sepecificity      = {round(confMatrix[0]/(confMatrix[0]+confMatrix[1]),4)}
""")
```

```python
rocArea = roc_auc_score(y_true = ytest,y_score = prediction)
```

```python
fpr ,tpr,threshHold = roc_curve(y_true = ytest,y_score = prediction,pos_label = 1)
```

```python
plt.plot(fpr,tpr,'o--')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.text(0.35,0.05,f'Area Under Curve = {rocArea:0.2f}',{'fontsize':'large'})
plt.savefig('./images/ROC.jpg')
plt.show()
```
